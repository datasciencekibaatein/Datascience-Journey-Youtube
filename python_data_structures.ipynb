# 📋 PART 1: LIST

A **list** is an ordered, mutable collection. It allows duplicate elements and supports indexing & slicing.

```python
# ── Creating Lists ──────────────────────────────────────────────────────────
lst = [23,45,23,34,89,90]
print(lst)
print(type(lst))
```

```python
# ── Indexing & Slicing ───────────────────────────────────────────────────────
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(fruits[1:3])
```

```python
# ── append() — Add one element at the end ───────────────────────────────────
fruits.append("Guava")
```

```python
print(fruits)
```

```python
# ── insert() — Insert at specific index ─────────────────────────────────────
fruits.insert(3,"water melon")
```

```python
print(fruits)
```

```python
# ── extend() — Add multiple elements ────────────────────────────────────────
temp = ["kivi","orange","melon"]
fruits.extend(temp)
print(fruits)
```

```python
# ── remove() — Remove by value (first occurrence) ───────────────────────────
fruits.remove("melon")
```

```python
fruits.remove("banana")
```

```python
print(fruits)
```

```python
# ── pop() — Remove by index (default: last) ──────────────────────────────────
fruits.pop()
```

```python
fruits.pop(2)
```

```python
print(fruits)
```

```python
# ── sort()  ──────────────────────────────────────────────────────
nums = [5, 2, 9, 1, 7, 3]

nums.sort()
```

```python
print(nums)
```

```python
nums = [5, 2, 9, 1, 7, 3]
nums.sort(reverse=True)
```

```python
print(nums)
```

```python
# ── index(), count(), len() ───────────────────────────────────────────────────
data = [10, 20, 30, 20, 40, 20]

data.index(40)
```

```python
# ── min(), max(), sum() ────────────────────────────────────────────────────────
min(data)
```

```python
max(data)
```

```python
sum(data)
```

# 📌 PART 2: TUPLE

A **tuple** is an ordered, **immutable** collection. Faster than lists and used for fixed data (e.g., coordinates, RGB values, DB records).

```python
# ── Creating Tuples ──────────────────────────────────────────────────────────
tp = (34,23,56,78)
print(tp)
print(type(tp))
```

```python
# ── Indexing & Slicing ───────────────────────────────────────────────────────
tp[3]
```

```python
# ── count() and index() ──────────────────────────────────────────────────────
tpp = (34,23,34,56,23,89,34)

print(tpp)

print(tpp.count(34))
print(tpp.index(34))
```

```python
# ── Tuple Unpacking ──────────────────────────────────────────────────────────

num1,num2,num3 = (23,45,67)

print(num1,num2,num3)
```

# 📌 PART 3: SET

A **set** is an unordered collection of **unique** elements. Ideal for membership testing and eliminating duplicates.

```python
# ── Creating Sets ────────────────────────────────────────────────────────────

st = {34,23,34,23,67,89}
print(st)
```

```python
# ── add() and update() ───────────────────────────────────────────────────────

st.add(56)
st.add(67)
print(st)

```

```python
st.update([23,45,98])
print(st)
```

```python
# ── remove() vs discard() ────────────────────────────────────────────────────
st.remove(90)
print(st)
```

```python
print(st)
```

```python
st.discard(45)
print(st)
```

```python
st.discard(90)
print(st)
```

```python
# ── SET OPERATIONS ───────────────────────────────────────────────────────────
batch_A = {"Alice", "Bob", "Carol", "David"}
batch_B = {"Carol", "David", "Eve", "Frank"}

# Union — all students
print(batch_A.union(batch_B))
# Intersection — common students
print(batch_B.intersection(batch_A))
# Difference — in A but NOT in B
print(batch_B.difference(batch_A))
```

```python
# ── issubset(), issuperset()────────────────────────────────────
a = {1, 2, 3}
b = {1, 2, 3, 4, 5}
c = {10, 20}

print(c.issubset(b))
print(b.issuperset(a))
```

# 📌 PART 4: DICTIONARY

A **dictionary** stores key-value pairs. Keys must be unique and hashable. Ordered from Python 3.7+.

```python
# ── Creating Dictionaries ─────────────────────────────────────────────────────

student = {
    "Name":"Rahul",
    "Age":20,
    "Marks":45
}
```

```python
print(type(student))
```

```python
# ── Accessing Values ──────────────────────────────────────────────────────────
student["Marks"]
```

```python
student.get("Name")
```

```python
print(student)
```

```python
# ── Adding & Updating ────────────────────────────────────────────────────────
student["Name"] = "Sohit"
student['Age'] = 20
student['City'] = "Delhi"
```

```python
print(student)
```

```python
# ── keys(), values(), items() ────────────────────────────────────────────────
student.keys()
```

```python
student.values()
```

```python
student.items()
```
