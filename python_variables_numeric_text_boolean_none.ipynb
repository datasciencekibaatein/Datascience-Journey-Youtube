# ЁЯУж PART 1 тАФ VARIABLES
## 1.1 What is a Variable?

A **variable** is a named label that points to a value stored in memory. Python does not require you to declare a variable's type тАФ it is inferred automatically at runtime.

```
variable_name = value
        тЖС              тЖС
    identifier      object in memory
```

## 1.2 Why use Variables?
- Store and reuse data without hardcoding
- Make code readable and maintainable
- Allow values to change over time

## 1.3 Naming Rules & Conventions

**Rules (must follow тАФ Python will error otherwise):**
- Must start with a **letter** or **underscore** `_`
- Can contain letters, digits, and underscores
- **Case-sensitive**: `score` тЙа `Score` тЙа `SCORE`
- Cannot be a Python **keyword** (`if`, `for`, `class`, `return`, etc.)

**Conventions (PEP 8 тАФ best practice):**

| Style | Used For | Example |
|---|---|---|
| `snake_case` | variables, functions | `student_name` |
| `UPPER_SNAKE_CASE` | constants | `MAX_RETRIES` |
| `PascalCase` | class names | `StudentProfile` |
| `_single_underscore` | internal/private | `_helper` |
| `__double_underscore` | name mangling | `__secret` |

## 1.4 Assignment тАФ How to Create Variables

````python
num = 34
````

## 1.5 Dynamic Typing

Python is **dynamically typed** тАФ a variable can be reassigned to a completely different type. The type is attached to the **value**, not the variable name.

````python
print(num)
num = 54
print(num)
````

## 1.6 How Python Stores Variables in Memory

In Python, variables are **references** (pointers) to objects. Two variables can point to the same object.

````python
# Deleting a variable
#del temp
# print(temp)   # NameError: name 'temp' is not defined

del num

print(num)
````

## 1.8 Constants

Python has no built-in constant type. By convention, constants are written in `ALL_CAPS` to signal they should not be changed.

````python
PI = 3.14
print(PI)
PI = 23
print(PI)
````

# ЁЯФв PART 2 тАФ NUMERIC TYPES
## 2.1 What are Numeric Types?

| Type | Description | Example |
|---|---|---|
| `int` | Whole numbers | `42`, `-7`, `0` |
| `float` | Decimal numbers | `3.14`, `-0.5` |
| `complex` | Real + imaginary | `3 + 4j` |

### `int` тАФ Integer

````python
age = 23

print(age)
print(type(age))
print(id(age))
````

### `float` тАФ Floating Point

````python
price = 45.34


print(price)
print(type(price))
print(id(price))
````

### `complex` тАФ Complex Numbers

````python
imng = 3 + 3j

print(imng)
print(type(imng))
print(id(imng))
````

### Type Conversion тАФ Numeric

````python
price = 34.6 
print(type(price))
amount = int(price)
print(amount)

age = 34
agee = float(34)
print(agee)

data = "34.5"

print(type(data))

data = int(float(data))
print(data)
````

## 3.1 What is a String?

A `str` is an **ordered, immutable** sequence of Unicode characters. Strings are the most used data type in real-world Python.

````python
name = 'Dhruv'
name = "Krish432"
print(name)
print(type(name))

multiLine = '''hi hello, how are you
where are you from'''

print(multiLine)
````

# тЬЕ PART 4 тАФ BOOLEAN TYPE: `bool`
## 4.1 What is Boolean?

`bool` has exactly two values: `True` and `False`. It is a **subclass of `int`** (`True == 1`, `False == 0`).

````python
flag = True
print(flag)
print(type(flag))
print(id(flag))


signal = False
print(signal)

````

# ЁЯЪл PART 5 тАФ NONE TYPE: `NoneType`
## 5.1 What is None?

`None` is a **singleton** of type `NoneType`. It represents the **intentional absence of a value** тАФ Python's equivalent of `null`.

````python
var = None
print(None)
print(type(None))
````

# ЁЯУЛ Master Quick Reference

| Concept | Key Rule |
|---|---|
| Variable | Named reference; no type declaration needed |
| `int` | Arbitrary precision whole numbers |
| `float` | IEEE 754 decimal; use `isclose()` not `==` |
| `complex` | `real + imagj`; `.real`, `.imag` attributes |
| `str` | Immutable Unicode sequence; sliceable |
| `bool` | Subclass of `int`; `True=1`, `False=0` |
| `None` | Singleton; check with `is`, never `==` |
