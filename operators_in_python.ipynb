# Operators in Python 

Operators are special symbols that perform operations on values (called **operands**). In the expression `5 + 3`, the symbol `+` is the operator and `5`, `3` are the operands.

Python supports **seven categories** of operators:

1. Arithmetic Operators
2. Assignment Operators
3. Comparison (Relational) Operators
4. Logical Operators
5. Identity Operators
6. Membership Operators
7. Bitwise Operators

By the end of this notebook you will know what each operator does, how it behaves on different data types, and where you will use it in real code.

---

## 1. Arithmetic Operators

Arithmetic operators perform basic mathematical operations.

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `7 + 2` | `9` |
| `-` | Subtraction | `7 - 2` | `5` |
| `*` | Multiplication | `7 * 2` | `14` |
| `/` | Division (float) | `7 / 2` | `3.5` |
| `//` | Floor Division | `7 // 2` | `3` |
| `%` | Modulus (remainder) | `7 % 2` | `1` |
| `**` | Exponent (power) | `7 ** 2` | `49` |

````python
num1 = 23
num2 = 34

res = 4 **  2
print(res)
````

---
## 2. Assignment Operators

The `=` operator assigns a value to a variable. The **compound** assignment operators perform an arithmetic or bitwise operation *and* assign the result back to the variable in one step.

| Operator | Equivalent to | Example |
|----------|---------------|---------|
| `=` | — | `x = 5` |
| `+=` | `x = x + y` | `x += 3` |
| `-=` | `x = x - y` | `x -= 3` |
| `*=` | `x = x * y` | `x *= 3` |
| `/=` | `x = x / y` | `x /= 3` |
| `//=` | `x = x // y` | `x //= 3` |
| `%=` | `x = x % y` | `x %= 3` |
| `**=` | `x = x ** y` | `x **= 3` |
| `&=` `\|=` `^=` `>>=` `<<=` | bitwise versions | see Section 7 |

````python
x = 3
print(x) # x = 3
x %= 5 # x = x % 5
print(x)
````

---
## 3. Comparison (Relational) Operators

Comparison operators compare two values and return a **Boolean** (`True` or `False`).

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<` | Less than | `5 < 3` | `False` |
| `>=` | Greater than or equal | `5 >= 5` | `True` |
| `<=` | Less than or equal | `5 <= 4` | `False` |

````python
age = 18

print(age <= 18)
````

### Common pitfall — `=` vs `==`

- `=` is **assignment** (puts a value into a variable).
- `==` is **comparison** (checks if two values are equal).

Mixing them up is one of the most common beginner bugs.

## 4. Logical Operators

Logical operators combine Boolean expressions.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `and` | True if **both** sides are True | `True and False` | `False` |
| `or` | True if **at least one** side is True | `True or False` | `True` |
| `not` | Inverts the Boolean value | `not True` | `False` |

````python
salary = 13000
age = 18

print(salary > 20000 or age >= 18)
````

````python
print(not salary > 15000)
````

## 5. Identity Operators

Identity operators check whether two variables point to the **same object in memory** (not just whether they have the same value).

| Operator | Meaning |
|----------|---------|
| `is` | True if both operands refer to the same object |
| `is not` | True if they refer to different objects |

````python
lst = [1,3,4,5,3,5,6]
temp = lst

re = [1,3,4,5,3,5,6]
print(lst is temp)
````

---
## 6. Membership Operators

Membership operators test whether a value is present inside a **sequence** (string, list, tuple, set, dict).

| Operator | Meaning |
|----------|---------|
| `in` | True if the value is found in the sequence |
| `not in` | True if the value is **not** found |

````python
print(lst)

name = "Dhruv verma"
print("verma" not in name)
````

## 7. Bitwise Operators

Bitwise operators work on the **binary representation** of integers. They are mostly used in low-level programming, flag handling, image processing, and competitive programming.

| Operator | Name | Example |
|----------|------|---------|
| `&` | AND | `5 & 3` → `1` |
| `\|` | OR | `5 \| 3` → `7` |
| `^` | XOR | `5 ^ 3` → `6` |
| `~` | NOT (one's complement) | `~5` → `-6` |
| `<<` | Left shift | `5 << 1` → `10` |
| `>>` | Right shift | `5 >> 1` → `2` |

**Decimal 5** in binary is `0101`. **Decimal 3** is `0011`.

```
  0101   (5)             0101   (5)             0101   (5)
& 0011   (3)           | 0011   (3)           ^ 0011   (3)
------                 ------                 ------
  0001   (1)             0111   (7)             0110   (6)
```

````python
print(5 ^ 3)
````

````python
print(~ 6) # -(n+1)
````

````python
print( 5 << 1)
````

````python
print(5 >> 1)
````
