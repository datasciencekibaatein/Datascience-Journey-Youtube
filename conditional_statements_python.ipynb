# Python Conditional Statements 


## 1️⃣ What Are Conditional Statements?

Conditional statements allow your program to **make decisions** based on whether a condition is `True` or `False`.

Think of it like real life:
> *If it is raining → carry an umbrella, otherwise → go out normally.*

In Python, every condition evaluates to a **Boolean**: `True` or `False`.

---

### 🔑 Syntax

```python
if condition:
    # block executed when condition is True
elif another_condition:
    # block executed when another_condition is True
else:
    # block executed when none of the above are True
```

**Key rules:**
- Indentation is mandatory (4 spaces or 1 tab)
- The colon `:` at the end of each clause is required
- `elif` and `else` are optional

## 2️⃣ Simple `if` Statement

````python
age = 16

if age>=18:
    print("You are allowed to vote")
````

````python
age>=18
````

````python
temperature = 33

if temperature>36:
    print("you are sick")
````

## 3️⃣ `if-else` Statement

Use `else` when you want to handle the case where the condition is **False**.

````python
temperature = 33

if temperature>36:
    print("you are sick")
else:
    print("you temerature is below 36")
````

## 4️⃣ `if-elif-else` Statement

Use `elif` (short for *else-if*) when you have **multiple conditions** to check in sequence.

Python checks each condition **top to bottom** and stops at the first `True` one.

````python
marks = 50

if marks >=90:
    print("Grade A")
elif marks >= 80:
    print("Grade B")
elif marks >=70:
    print("Grade C")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
````

## 5️⃣ Comparison Operators

These operators compare two values and return `True` or `False`.

| Operator | Meaning | Example |
|----------|---------|--------|
| `==` | Equal to | `5 == 5` → `True` |
| `!=` | Not equal to | `5 != 3` → `True` |
| `>` | Greater than | `7 > 3` → `True` |
| `<` | Less than | `3 < 7` → `True` |
| `>=` | Greater than or equal | `5 >= 5` → `True` |
| `<=` | Less than or equal | `4 <= 5` → `True` |

## 6️⃣ Logical Operators

Combine multiple conditions using logical operators.

| Operator | Description | Example |
|----------|-------------|--------|
| `and` | True if **both** conditions are True | `x > 0 and x < 10` |
| `or` | True if **at least one** condition is True | `x < 0 or x > 100` |
| `not` | **Reverses** the Boolean value | `not (x == 5)` |

````python
salary = 20000
age = 18

if age > 18 or salary > 20000:
    print("you are allowed")
else:
    print("Not allowed")
````

````python
age > 18
````

````python
if not age > 18:
    print("you are eligible")
````

## 7️⃣ Nested Conditional Statements

A conditional statement **inside** another conditional statement is called **nested conditionals**.

> ⚠️ Keep nesting to a maximum of 2–3 levels for readability.

````python
username = "dhru"
password = "1234"

if username =="dhruv":
    if password == "12345":
        print("you are logged in!")
    else:
        print("password is not correct")

else:
    print("username is not correct")
````

## 8️⃣ Ternary (Inline) Conditional Expression

Python supports a **one-line if-else** called a **ternary operator** or **conditional expression**.

```python
value = <true_value> if <condition> else <false_value>
```

Use it for **simple assignments** — avoid it for complex logic.

````python
num = 37

print("even") if num % 2==0 else print("odd")
````

````python
num % 2
````

## 9️⃣ `match-case` Statement (Python 3.10+)

Python 3.10 introduced `match-case`, similar to `switch` statements in other languages.

It is ideal when comparing a variable against **multiple fixed values**.

````python
week_num = 6

match week_num:
    case 1:
        print("sunday")
    case 2:
        print("monday")
    case 3:
        print("tuesday")
    case 4:
        print("wednesday")
    case 5:
        print("thursday")
    case 6:
        print("friday")
    case 7:
        print("saturday")
````
