# Strings in Python

A **string** is a sequence of characters used to store and represent text. Strings are one of the most used data types in Python — for names, messages, file paths, JSON, HTML, and almost any text you process.


````python
name = 'dhruv'

print(name)
print(type(name))
````

## 1. Creating Strings



**Tip:** Use double quotes when the text contains an apostrophe, and single quotes when it contains double quotes — this avoids extra escaping.

````python
first_name = "dhruv"
last_name = 'verma'

print(first_name)
print(last_name)

msg = ''' hello  how are you ?
you did not msg me 
where were you ?'''

print(msg)
````

````python
status = "it's raining"

print(status)
````

````python
msg = 'i am learning "Python".'
print(msg)
````

## 2. Strings Are Immutable

Once created, a string **cannot be changed in place**. Any operation that "modifies" a string actually returns a **new** string. Trying to assign to a single character raises an error.


````python
txt = "hello there"
txt[1]='j'
````

## 3. Indexing

Each character has a position (**index**). Indexing starts at **0** from the left, and **-1** from the right (negative indexing).

```
 P   y   t   h   o   n
 0   1   2   3   4   5     <- positive index
-6  -5  -4  -3  -2  -1     <- negative index
```


````python
txt = "Python"

print(txt[-3])
````

## 4. Slicing

Slicing extracts a **substring** using the syntax `string[start:stop:step]`.

- `start` — index to begin (inclusive). Defaults to the beginning.
- `stop` — index to end (**exclusive**). Defaults to the end.
- `step` — how many characters to jump. Defaults to `1`.


````python
txt = "python"

print(txt[0:5:3])
print(txt[0::2])
print(txt[::-1])
````

## 5. Concatenation & Repetition

Use `+` to join strings together and `*` to repeat a string.


````python
name = "dhruv "
last_name = "verma"

print(name + last_name)
print("=" * 20)
````

## 6. Length & Membership

- `len()` returns the number of characters.
- The `in` operator checks whether a substring exists.


````python
msg = "python"

print(len(msg))
````

````python
txt = "i am learning python"

print("java" in txt)
````

## 7. Common String Methods

Strings come with many built-in methods. Remember: they **return a new string** and never change the original.

| Method | What it does |
|---|---|
| `.upper()` / `.lower()` | change case |
| `.title()` / `.capitalize()` | title case / first letter |
| `.strip()` | remove surrounding whitespace |
| `.replace(a, b)` | replace `a` with `b` |
| `.find(x)` | index of first match (`-1` if none) |
| `.count(x)` | how many times `x` appears |
| `.startswith()` / `.endswith()` | boolean checks |


````python
name = "dhruv"
print(name.upper())
last_name = "VERMA"
print(last_name.lower())
````

````python
lang = "python is good programming language"

print(lang.title())
print(lang.capitalize())
````

````python
msg = "     i am learning java.     "

print(len(msg))
print(len(msg.strip()))
print(msg.count("i"))
````

````python
filename = "firstcode.py"

print(filename.endswith(".py"))
print(filename.startswith("first"))
````

````python
msg = "i am learning python"

print(msg.find("java"))
print(msg.replace("python","java"))
````

Methods can be **chained** because each returns a new string:

````python
msg = "    i am learning python.  "
new_msg = msg.strip()
print(new_msg.replace("python","java"))
````

## 8. `split()` and `join()`

- `.split(sep)` breaks a string into a **list** of pieces.
- `sep.join(list)` does the opposite — combines a list into one string.


````python
data = "rahul,23,delhi,IT"

print(data.split(","))

lst = data.split(",")
````

````python
" | ".join(lst)
````

## 9. Looping Over a String

A string is iterable, so you can loop through it character by character.


````python
for ch in "python":
    print(ch)
````

## 10. Checking String Content

These boolean methods are handy for validating user input.

| Method | Returns `True` if the string... |
|---|---|
| `.isdigit()` | contains only digits |
| `.isalpha()` | contains only letters |
| `.isalnum()` | contains only letters/digits |
| `.isspace()` | contains only whitespace |
| `.islower()` / `.isupper()` | is all lower/upper case |


````python
num = "123"
print(num.isdigit())
alnum = '123hello'
print(alnum.isalnum())
name = "dhruv"
print(name.isalpha())

sp = "   "
print(sp.isspace())

print(name.isupper())

l_name = "VERMA"
print(l_name.isupper())
````
