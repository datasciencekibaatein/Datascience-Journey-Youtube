# What are loops ?

A loop is a control-flow statement that repeats a block of code either a fixed number of times or until a condition becomes false.

## Types of loops

- for loop
- while loop

<h2 style="color:green;"><strong>for loop:</strong></h2>

### for loop with datastructure

```python
nums = [2,3,5,2,6,7,8]
for num in nums:
    print(num * 2)
```

```python
tp = (35,23,90,34)

for t in tp:
    print(t)
```

```python
student = {
    "name":"Ram",
    "Age":23,
    "course":"Machine learning"
}

for k,v in student.items():
    print("Key:",k," value: ",v)
```

### for loop with range

range(stop)

range(start,stop)

range(start,stop,step)

```python
for i in range(9):
    print(i)
```

```python
for i in range(10,20):
    print(i)
```

```python
for i in range(2,22,3):
    print(i)
```

### for loop with enumerate() function

```python
lst = ["Rahul","Krish","Sohit","Krishna"]

for index,name in enumerate(lst):
    print(index,": ", name)
```

### for loop with zip() function

```python
a = [34,23,45,653,76]
b = [78,23,23,5,90]
c = [45,67,78,34,87,78]

for aa,bb,cc in zip(a,b,c):
    print(aa," ", bb ," ", cc)
```

### List comprehension

```python
for i in range(2,10):
    if i % 2 ==0:
        print(i)
```

```python
lt = [0,1,2,3,4,5]
```

```python
temp = [i for i in range(21) if i % 2==0]

print(temp)
```

### Loop control - break,continue

```python
for i in range(2,10):
    if i == 7:
        break
    print(i)
```

```python
for i in range(2,10):
    if i == 5:
        continue
    print(i)
```

## for loop with string

```python
name = "dhruv"
for s in name:
    print(s)
```

<h2 style="color:green;"><strong>while loop</strong></h2>

```python
i = 0

print(5<5)
```

```python
i = 1

while i<=10:
    print(i*2)
    i = i + 1
```

## while with - break,continue

```python
i = 1

while i<=10:
    if i == 5:
        break

    print(i)
    i = i + 1
```

```python
i = 1

while i<=10:
    
    i = i + 1
    if i == 5:
        continue

    print(i)
    
```

<h2 style="color:green;"><strong>Nested Loops</strong></h2>

```python
for i in range(2,5):
    for j in range(2,5):
        print(i,j)
```
